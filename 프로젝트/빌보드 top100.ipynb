```python
import bs4,requests

billboard = "https://www.billboard.com/charts/hot-100/"

print("빌보드 스크레이핑...")
res = requests.get(billboard)
res.raise_for_status()
soup = bs4.BeautifulSoup(res.text,"html.parser")
```

find_all(’태그’{‘class혹은 id’:’class명 또는 id명’   ,   ‘class혹은 id’:’class명 또는 id명’   ,   여러개 가능!  })
특정 가능 한 태그 사용할것

```python
title_1=soup.find_all('h3',{'id':'title-of-a-story','class':'u-letter-spacing-0028@tablet'})
```

```python
title = soup.find_all('h3',{'id':'title-of-a-story','class':'lrv-u-font-size-18@tablet'})
```

```python
song_by = soup.find_all('span',{'class':'c-label','class':'a-no-trucate'})
```

```python
print(1," : ",title_1[0].getText().strip()  ," , ",  song_by[0].getText().strip())
for i in range(1,len(title)):
    a= title[i-1].getText()
    b=song_by[i].getText()

    print(i+1 , " : " ,a.strip()," , ",b.strip())

print(100," : ",title[-1].getText().strip()  ," , ",  song_by[-1].getText().strip())
```
