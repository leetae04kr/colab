```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
datasetfile="/content/lotto.csv"
df=pd.read_csv(datasetfile, encoding='utf-8')
df.head()
```

```python
df.info
```

# -----<nan>

#전처리

```python
numdf=df[['1','2','3','4','5','6']]
numdf=numdf.values.tolist()



numalldf=[]
for i in range(0,1044):
  numalldf.extend(numdf[i])
numalldf
```

```python
counter=[]
for i in range(1,1044*6+1):
  counter.append(i)
#counter
```

```python
numalldfpd = pd.DataFrame (numalldf, columns = ['numalldf'])
```

# -----<nan>

# 그래프

```python
import matplotlib.pyplot as plt
plt.scatter(counter,numalldf,marker='.', s=1, c='red')

```
