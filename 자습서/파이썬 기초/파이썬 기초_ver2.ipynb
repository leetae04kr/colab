함수에는 들어가는 놈 (input)이 있고, 나오는 놈 (output 혹은 return)이 있습니다. 정해진 로직(규칙)에 따라, input -> output으로 효율적으로 바꿔주는 역할을 합니다

````
a = 1
b = 2
c = 3
````

````
# 코드를 입력해 주세요
(a+b)*c
````

**[함수로 변경 후]**

````
def func(a, b, c):
    return (a + b) * c
````

````
# 코드를 입력해 주세요
func(1,2,3)
````

````
# 코드를 입력해 주세요
func(2,2,3)
````

````
# 코드를 입력해 주세요
func(3,2,3)
````

## def는 함수를 위한 정의 (define)
def 함수이름 (parameter1, parameter2, parameter...):

parameter는 함수로 부터 넘겨 받은 변수 또는 값 입니다.

끝에 콜론 (:) 빼먹지 않음에 주의 하세요!!

````
def my_func(var1):
    print(var1)
````

````
# 코드를 입력해 주세요
my_func('테디입니다! 반갑습니다')
````

## 함수는 값을 리턴할 수 있고, 안해도 됩니다

리턴이 없는 경우

````
def my_func(a, b):
    print(a, b)
````

````
# 코드를 입력해 주세요
my_func(1,10)
````

리턴이 있는 경우

````
def my_func(a, b):
    s = a + b
    return s
````

result라는 변수에 함수에 대한 리턴 값을 담아주세요

````
# 코드를 입력해 주세요
result= my_func(5,6)
````

````
result
````

리턴이 있는 경우는 변수에 값을 다시 할당 할 수 있습니다.

````
ret = my_func(1, 10)
````

````
print(ret)
````

````
print(ret + 10)
````
