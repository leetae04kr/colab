## 10번
$10.\ 좌표평면\ 위를\ 움직이는\ 점\ P의\ 시각\ t\ (t>0)에서의\ 위치\ (x,\ y)가\\
x=t-\frac{2}{t},\ y=2t+\frac{1}{t}\\
이다.\ 시각\ t=1에서\ 점\ P의\ 속력은?$

```python
from mpmath import *
fx = lambda t: t-2/t
fy = lambda t: 2*t+1/t
dfx = diff(fx, 1)
dfy = diff(fy, 1)
result = root(dfx**2 + dfy**2, n=2)
print(result, "= sqrt(10)")
```

## 15번
$15.\ 곡선\ y=2e^{-x}\ 위의\ 점\ P(t,\ 2e^{-t})\ (t\gt0)에서\ y축에\ 내린\ 수선의\ 발을\ A라\ 하고,\\
점\ P에서의\ 접선이\ y축과\ 만나는\ 점을\ B라\ 하자.\\
삼각형\ APB의\ 넓이가\ 최대가\ 되로록\ 하는\ t의\ 값은?$
![15번 문제](https://github.com/serithemage/2017_CSAT_Mathematics_Type_GA/blob/master/img/15.png?raw=1)
>P에서 접선을 구하여 B를 구하여 AB의 길이를 구한 뒤, AP의 길이 t와 곱한 값을 미분하여 0이 되는 값을 찾는다.

```python
from scipy import e
from sympy.solvers import solve
from sympy import Symbol, Limit, Derivative

t = Symbol('t')
AP = t
AB = 2 * t * e**(-t)

s = AP * AB / 2
d = Derivative(s, t).doit()
solve(d, t)
```

##17번
17. 좌표평면 위의 한 점 $(x,\ y)$에서 세 점 $(x+1,\ y),\ (x,\ y+1),\ (x+1,\ y+1)$ 중 한 점으로 이동하는 것을 점프라 하자.  
점프를 반복하여 점 $(0, 0)$에서 점 $(4, 3)$까지 이동하는 모든 경우 중에서, 임의로 한 경우를 선택할 때 나오는 점프의 횟수를 확률변수 X라 하자.  
다음은 확률변수 X의 평균 $E(X)$를 구하는 과정이다. (단, 각 경우가 선택되는 확률은 동일하다.)  
![17번 문제](https://github.com/serithemage/2017_CSAT_Mathematics_Type_GA/blob/master/img/17.png?raw=1)  
위의 (가), (나), (다)에 알맞은 수를 각각 a, b, c라 할 때, a+b+c의 값은?  

>(가): 확률변수 X가 가질 수 있는 값 중 가장 작은 값을 k라 하면 k는 대각선으로 3번, 가로로 1번 가는 경우인 4의 값을 갖는다.  
>(나): 대각선으로 1번, 가로로 3번, 위로 2번 가야한다.  
>(다): 위의 확률을 다 더해서 1이 나와야 한다.

```python
from sympy.solvers import solve
from sympy import Symbol

def factorial (item):
    if(item == 0):
        return(1)
    else:
        return (item * factorial(item-1))
    
a = 4
b = factorial(6)/(factorial(3)*factorial(2))
N = Symbol("N")
c = solve((4+30+b+35)/N -1, N)
print(a+b+float(c[0]))
```

##17번스크린샷 2022-11-29 오후 12.27.53.png

```python
import random
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

def dist(P, A, B):
    x2 = (A.x - P.x) * (B.y - P.y) - (A.y - P.y) * (B.x - P.x) 
    area = abs(x2)
    AB = ( (A.x - B.x) ** 2 + (A.y - B.y) ** 2 ) ** 0.5
    return ( area / AB )
n = 100000
s = 0
for i in range(0, n):
    x = 0
    y = 0
    for i in range(1,16):
        p = random.randrange(1,7)
        if p <= 2:
            x += 3
        if p >= 3:
            y += 1
    p = Point(x,y)
    a = Point(0,0)
    b = Point(1,-3/4)
    r = dist(p,a,b)
    s += r

print(s, s/n)
```

```python
from itertools import permutations

a = [1,2,3,4,5,6]
permute = permutations(a, 6)
ps = list(permute)
print("전체 순열",len(ps))
print("원 순열", len(ps) / 6)
count = 0
for p in ps:
    for i in range(0,6):
        if p[i] == 1:
            if i == 0:
                l = p[5]
                ll = p[4]
            elif i == 1:
                l = p[0]
                ll = p[4]
            else:
                l = p[i-1]
                ll = p[i-2]
            if i == 5:
                r = p[0]
                rr = p[1]
            elif i == 4:
                r = p[5]
                rr = p[0]
            else:
                r = p[i+1]
                rr = p[i+2]
            if l == 2 and ll != 3:
                count += 1
                print(p)
            if r == 2 and rr != 3:
                count += 1
                print(p)
print("1과 2가 양 사이드 일 경우", count / 6)
```

```python
from itertools import combinations_with_replacement

r = 0
for blackArray in combinations_with_replacement(["A", "B", "C", "D"], 6):
    for whiteArray in combinations_with_replacement(["A", "B", "C", "D"], 6):
        ab = blackArray.count("A")
        bb = blackArray.count("B")
        cb = blackArray.count("C")
        db = blackArray.count("D")
        aw = whiteArray.count("A")
        bw = whiteArray.count("B")
        cw = whiteArray.count("C")
        dw = whiteArray.count("D")
        if (bb + bw) > 0 and (cb +cw) > 0 and (db + dw) > 0:
            if ab >= 4 and ab > aw:
                moreBlack = 0
                if bb > bw:
                    moreBlack += 1
                if cb > cw:
                    moreBlack += 1
                if db > dw:
                    moreBlack += 1
                if moreBlack == 1:
                    r += 1
                    # print(blackArray, whiteArray)
                    # print(ab, bb, cb, db)
                    # print(aw, bw, cw, dw)

print("result", r)
```
