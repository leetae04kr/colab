```python
import random

def make_groups(member_list, group_size):
    random.shuffle(member_list)  # 그룹을 만들기 전에 리스트를 무작위로 섞음
    num_members = len(member_list)
    num_groups = -(-num_members // group_size)  # 올림을 이용하여 총 그룹 수를 계산

    groups = []
    for i in range(num_groups):
        start_idx = i * group_size
        end_idx = (i + 1) * group_size
        groups.append(member_list[start_idx:end_idx])

    return groups

def main():
    group = [ " 강혁주 " ," 안석규 " ," 홍은경 " ," 이산하 " ," 박해솔 " ," 배승아 " ," 배성원 " ," 이주영 " ," 조민성 " ," 이상연 " ," 김민재 " ," 어울림 " ," 이태영 " ," 권나현 " ," 윤수민 " ," 이주윤 " ," 양민주 " ," 이지연 " ," 이지은 " ," 김나홍 " ," 김채운 " ," 이승준 " ," 오욱진 " ," 정시윤 " ," 성민준 " ," 정예은 " ," 마지원 " ," 노현아 " ," 여동원 "  ," 오태경 " ," 이다윤 " ," 신영원 " ," 장호진 " ," 이주현 " ," 안보록 " ," 정승혜 " ," 이아린 " ," 신주연 " ," 장선민 " ," 현명빈 " ," 이연주 "]
    group_size = int(input("한 그룹 당 몇 명의 인원을 원하시나요? "))
    groups = make_groups(group, group_size)

    for i, group_members in enumerate(groups):
        print(f"group{i+1}={group_members}")

if __name__ == "__main__":
    main()
```
